```python
def calculate_fuels_fuel_cost(fuel_amounts, total_fuel_amounts, initial_fuel):
    """
    calculate the amount of fuel required to lift the amount of fuel provided.
    """
    new_fuel_amounts = np.where(fuel_amounts>=0, np.floor(fuel_amounts/3) -2, fuel_amounts)
    print(new_fuel_amounts)
    total_fuel_amounts = np.where(new_fuel_amounts>=0, total_fuel_amounts + new_fuel_amounts, total_fuel_amounts)
    print(total_fuel_amounts)
    
    if np.sum(new_fuel_amounts) < 0:
        print(np.sum(total_fuel_amounts))
        return total_fuel_amounts
    
    
    calculate_fuels_fuel_cost(new_fuel_amounts, total_fuel_amounts, initial_fuel)
```
